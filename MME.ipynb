## SUMO MME

```python
fp1 = sorted(glob.glob('/glade/derecho/scratch/wchapman/archive/supermodel/super_cam5/*.nc'))
fp2 = sorted(glob.glob('/glade/derecho/scratch/wchapman/archive/supermodel/super_cam6/*.nc'))
fpout = '/glade/derecho/scratch/wchapman/archive/supermodel/MME_56/'

for bingo in zip(fp1,fp2):
    DS1 = xr.open_dataset(bingo[0])
    DS2 = xr.open_dataset(bingo[1])
    
    vlist = ['U','V','T','Q','PRECC','PRECL','SWCF','LWCF','Z3','TREFHT','FLUT','TS','PSL','PS', 'U10']
    
    for varvar in vlist: 
        DS1[varvar][:] = (DS1[varvar].values+DS2[varvar].values)/2
        
    fout = f'{fpout}/MME.h0{bingo[0].split("h0")[-1]}'
    print(f'file out: {fout}')
    DS1.to_netcdf(fout,format='NETCDF4_CLASSIC')

FNS = sorted(glob.glob('/glade/derecho/scratch/wchapman/archive/supermodel/MME_56/MME.h0.*.nc'))
for FN in FNS: 
    print(FN)
    tparts = FN.split('.h0.')
    newname = tparts[0]+'_Combined.cam.h0.'+tparts[1]
    os.rename(FN,newname)
```

```python
FNS = sorted(glob.glob('/glade/derecho/scratch/wchapman/archive/supermodel/MME_56/MME_Combined*.nc'))
for FN in FNS: 
    print(FN)
    tparts = FN.split('.cam.')
    newname = tparts[0]+'.cam.h0.'+tparts[1]
    print(newname)
    # break
    os.rename(FN,newname)
```
